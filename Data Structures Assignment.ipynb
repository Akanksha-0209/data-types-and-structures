1. What are data structures, and why are they important?

Ans - Data Structures are methods for organizing and storing data effectively. They are important because they help  in managing and retrieving data in most efficient manner, making programs faster and easier to manage.

2. Explain the difference between mutable and immutable data types with examples?

Ans - Mutable Data types - These are the data types which can be modified or changed after they are created.
For example : List, dict, set
Immutable Data types - These are data types which cannot be changed after creation.
For example : Tuple

3. What are the main differences between lists and tuples in Python?

Ans - Lists - Lists are Mutable data structure.
They use more memory.
Example :  mylist = [1,2,3,4]

Tuples - Tuples are Immutable type of data structure.
They use less memory.
Example : mytuple = (1,2,3,4)
